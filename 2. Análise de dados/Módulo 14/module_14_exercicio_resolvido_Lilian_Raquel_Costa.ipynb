## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns
import pandas as pandas
import matplotlib.pyplot as plt

data_1 = sns.load_dataset("diamonds")
data_1.head()
```

```
# gráfico do exercício 1

diamonds = sns.barplot(data=data_1, x='cut', y='price', edgecolor='k', palette = 'pastel', errorbar = None)
diamonds.set(ylim = [0,5000]);
     
```

**Insight do gráfico 1**: O dataset apresenta uma característica chamada "tipo de corte". Esse parâmetro, aparentemente, apresenta alguma relação com o valor total do diamante. Os diamentes com corte Premium são mais caros e os de menor valor são os de corte ideal

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data_2 = sns.load_dataset("flights")
data_2.head()
```

```
import numpy as np

flights = data_2.query('month == "Dec"')
```

```
# gráfico do exercício 2

with sns.axes_style('whitegrid'):
  grafico_flights = sns.lineplot(x='year', y='passengers', data=flights);
  grafico_flights.set(title = 'Número de passageiros em dezembro por ano',
                      xlabel = 'Ano',
                      ylabel = 'Passageiros',
                      ylim = [100, 500]);
  grafico_flights.grid(alpha=0.35)     
  plt.xticks(np.arange(1949,1961,1))                 
```

**Insight do gráfico 2**: A quantidade de passegeiros em dezembro tem uma tendência clara de aumento ao longo dos anos. Entre os anos 1957-1958 existe uma tendência contrária, onde o número de passageiros foi igual para o mês de dezembro de cada ano. Entre 1952 e 1953 foi quando o mês de dezembro sofreu a menor variação, quando comparado com os demais anos, ainda que essa variação tenha sido de crescimento.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data_3 = sns.load_dataset("flights")
data_3.head()
```

```
flights_month = data_3.query('year >= 1949 and year <= 1959')
```

```
# gráfico do exercício 3

with sns.axes_style('whitegrid'):
  grafico_flights_month = sns.lineplot(x='month', y='passengers', hue='year', data=flights_month, palette = 'pastel');
  grafico_flights_month.set(title = 'Número de passageiros por mês',
                      xlabel = 'Mês',
                      ylabel = 'Passageiros')    
  grafico_flights_month.grid(alpha=0.35)     
  grafico_flights_month.legend(title='Ano',loc=[1.01,0]) 
```

**Insight do gráfico 3**: Os meses de janeiro, fevereiro, maio e novembro são meses com menor quantidade de voos. Já os meses de março, agosto e dezembro são meses com maior quantitativo de voos. No decorrer do tempo, os dados apresentam uma mesma tendência quanto ao número de passageiros por mês.
